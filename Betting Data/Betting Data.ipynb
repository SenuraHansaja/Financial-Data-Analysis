```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive',force_remount=True)
```

```python
file = "drive//MyDrive//RAW//superbowl_all.csv"
data = pd.read_csv(file).set_index("Unnamed: 0")
data.info()
```

```python
#@title 1st Half Money Line DataFrame
##creating a mask for the first half moneyline
money_line  = data[data["market_name"] == "1st Half Moneyline"]
money_line.info()
```

```python
money_line.head()
```

```python
two_sided = pd.DataFrame(money_line)
two_sided.drop(columns=["id","checked_date","price","sports_book_name"],inplace=True)
```

```python
two_sided.info()
```

```python
two_sided_moneyline = pd.DataFrame()
for column in two_sided.columns:
  val = [i for i in two_sided[column].unique()]
  if len(val) <=1:
    val.append(val[0])
  two_sided_moneyline[column] = val
```

```python
two_sided_moneyline.head()
```

```python
###Kansas City Chiefs mask to get the betting centers which allow betting the team
KansasCityChiefs = money_line[money_line["name"] == "Kansas City Chiefs"]
PhiladelphiaEagles	 = money_line[money_line["name"] == "Philadelphia Eagles"]
KansasCityChiefs[0:10]
```

```python
sports_book_names1 = ["sports_book_name_"+str(i+1) for i in range(len(KansasCityChiefs['sports_book_name']))]
```

```python
books_dict = {}
for (a, b, c) in zip(sports_book_names1,KansasCityChiefs['sports_book_name'],PhiladelphiaEagles["sports_book_name"]):
  books_dict[a] = [b,c]
```

```python
books_DataFrame = pd.DataFrame(books_dict)
two_sided_moneyline = pd.concat([two_sided_moneyline,books_DataFrame], axis=1)
```

```python
two_sided_moneyline
```

```python
#@title Appending the odd values
odd_names = ["odd_"+str(i+1) for i in range(len(range(len(KansasCityChiefs['price']))))]
odd_dict = {}
for (a,b,c) in zip(odd_names,KansasCityChiefs['price'],PhiladelphiaEagles['price']):
  odd_dict[a] = [b,c]
```

```python
odd_DataFrame = pd.DataFrame(odd_dict)
two_sided_moneyline = pd.concat([two_sided_moneyline, odd_DataFrame], axis=1)
two_sided_moneyline.info()
```

```python
#@title to earn profit we do the market arbitrageur
## reverse of the betting center is the hedge book
hedge_names = ["hedge_book_"+str(i+1) for i in range(len(range(len(KansasCityChiefs['name']))))]
hedge_book = {}
for (a,b,c) in zip(hedge_names,KansasCityChiefs['sports_book_name'],PhiladelphiaEagles['sports_book_name']):
  hedge_book[a] = [c,b]
```

```python
hedge_DataFrame = pd.DataFrame(hedge_book)
two_sided_moneyline = pd.concat([two_sided_moneyline, hedge_DataFrame], axis=1)
two_sided_moneyline
```

```python
#@title Reverse Bet using the market arbitrageur
hedge_odd = ["hedge_odd_"+str(i+1) for i in range(len(range(len(KansasCityChiefs['price']))))]
hedge_odd_dict = {}
for (a,b,c) in zip(hedge_odd,KansasCityChiefs['price'],PhiladelphiaEagles['price']):
  hedge_odd_dict[a] = [c,b]

hedge_odd_DataFrame = pd.DataFrame(hedge_odd_dict)
two_sided_moneyline = pd.concat([two_sided_moneyline, hedge_odd_DataFrame], axis=1)
two_sided_moneyline
```
